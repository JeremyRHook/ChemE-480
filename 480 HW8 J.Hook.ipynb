6. Simulate the results of your controllers for a unit step distubance using the control module

In a single plot compare (1) and (2) using FF control alone

```python
#Found Parameters From Problem 1
theta = 0.5
den = [4, 5, 1]
Gd = tf([2], den)
Gp = tf([1], [1, 1])
Gc = tf([1,1], [3,0])
Gt = tf(*pade(theta, n = 3))
Gm = 1
Gv = 1

#Time 
ts = np.linspace(0, 20, 500)
```

```python
# Steady State feedforward controller
SSFF = Gd - tf([8, 2], den)*Gt

# Dynamic Analysis feedforward controller
DynamicFF = Gd - tf([1,2], den)*Gt

SS, t = step(SSFF, ts)
dynamic, t = step(DynamicFF, ts)
```

```python
plt.figure(1)
plt.plot(t,SS, label='Steady State FeedForward')
plt.plot(t,dynamic, label='Dynamic Analysis FeedForward')
plt.xlabel('Time [Seconds]') 
plt.ylabel('Controller Output')
plt.ylim(top=2, bottom=-2)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.title('Steady State, Dynamic Analysis FeedForward Control')

print('In a single plot compare (1) and (2) using FF control alone')
```

In a separate plot compare (3), (4), and (5)

```python
# FeedBack controller
FB = Gd / (1 + Gc*Gv*Gp*Gm)

# Steady State FeedBack controller
SS_FF_FB = tf([6,0], den) - tf([24, 6, 0], den)*Gt

# Dynamic Analysis FeedForward + FeedBack controller
Dynamic_FF_FB = tf([6, 0], den) - tf([3, 6, 0], den)*Gt

FB, t = step(FB, ts)
SS_FB, t = step(SS_FF_FB, ts)
Dynamic_FB, t = step(Dynamic_FF_FB, ts)
```

```python
plt.figure(1)
plt.plot(t,FB,label='FeedBack Controller')
plt.plot(t,SS_FB, label = 'Steady State & FeedBack Controller')
plt.plot(t,Dynamic_FB, label ='Dynamic Analysis & FeedBack Controller')
plt.xlabel('Time [Seconds]') 
plt.ylabel('Controller Output')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.ylim(top=2, bottom=-2)
plt.title('Comparison of Different Controllers')

print('In a separate plot compare (3), (4), and (5)')
```
